```python
import pandas as pd
import requests as req
import re

#https://soundcloud.com/discover
url=input('URL : ')
url_s1=[]
url_s2=[]
url_s=[]
url_type=[]
low=1
lows=[]
r=req.get(url,timeout=3)
text1=r.text.split('\n')
for i in text1:
  url_s1=re.findall('"https://(.*?)"',i)
  url_s+=url_s1
for i in url_s:
  r=req.get('https://'+i,timeout=3)
  url_s1=re.findall('"https://(.*?)"',r.text)
  url_s2+=url_s1
url_s+=url_s2
for j in url_s:
  if ('.png' or '.jpg' or '.webp' or '.svg' or '.ico' or '.jpeg') in j:
    url_type.append('img')
  elif ('.js') in j:
    url_type.append('javascript')
  elif ('.css') in j:
    url_type.append('css')
  elif ('.mp3' or '.wav' or 'weba') in j:
    url_type.append('audio')
  elif ('.mp4' or '.avi' or '.webm') in j:
    url_type.append('video')
  elif ('.woff' or '.woff2') in j:
    url_type.append('font')
  else:
    url_type.append(float('nan'))
df=pd.DataFrame(url_s)
df[1]=pd.DataFrame(url_type)
df.columns=['URL_list','URL_type']
df.sort_values('URL_type')
```
