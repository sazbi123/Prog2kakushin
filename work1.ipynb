```python
import pandas as pd
import requests as req
#https://soundcloud.com/discover
url=input('URL : ')
url_s=[]
url_type=[]
low=1
lows=[]
r=req.get(url,timeout=3)
text1=r.text.split('\n')
for i in text1:
  if 'http' in i:
    url_s.append(i)
    lows.append(low)
  low+=1
for j in url_s:
  if ('.png' or '.jpg') in j:
    url_type.append('img')
  elif ('.js') in j:
    url_type.append('javascript')
  else:
    url_type.append('None')
df=pd.DataFrame(lows)
df[1]=pd.DataFrame(url_s)
df[2]=pd.DataFrame(url_type)
df.columns=['low_number','URL_list','URL_type']
df
```
