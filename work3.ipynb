```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img=cv2.imread('/content/7e6a130e-f56b-4739-9442-270be80acf76.jpg')
h = img.shape[0]
w = img.shape[1]
def xor(img,h,w):
  for iy in range(h):
    for ix in range(w):
      img[iy,ix][2]=int(bin(~img[iy,ix][2]),base=2)
      img[iy,ix][1]=int(bin(~img[iy,ix][1]),base=2)
      img[iy,ix][0]=int(bin(~img[iy,ix][0]),base=2)
  return img
def r_only(img,h,w):
  for iy in range(h):
    for ix in range(w):
      img[iy,ix][1]=0
      img[iy,ix][0]=0
  return img
def g_only(img,h,w):
  for iy in range(h):
    for ix in range(w):
      img[iy,ix][2]=0
      img[iy,ix][0]=0
  return img
def b_only(img,h,w):
  for iy in range(h):
    for ix in range(w):
      img[iy,ix][2]=0
      img[iy,ix][1]=0
  return img
def left_rotate(img,h,w,count):
  for co in range(count):
    for iy in range(h):
      for ix in range(w):
        img[iy,ix][2]=int(bin(img[iy,ix][2]<<1),base=2)+int(bin(img[iy,ix][2])[2],base=2)
        img[iy,ix][1]=int(bin(img[iy,ix][1]<<1),base=2)+int(bin(img[iy,ix][2])[2],base=2)
        img[iy,ix][0]=int(bin(img[iy,ix][0]<<1),base=2)+int(bin(img[iy,ix][2])[2],base=2)
  return img
def r_bit(img,h,w,num):
  img2 = np.zeros( (h, w), np.uint8 )
  for iy in range(h):
    for ix in range(w):
      try:
        if (bin(img[iy,ix][2])[len(bin(img[iy,ix][2]))-num-1])=='1':
          img2[iy,ix]=255
      except IndexError:
        pass
  return img2
cv2_imshow(r_bit(img,h,w,0))
```
