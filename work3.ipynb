```python
import cv2
from google.colab.patches import cv2_imshow
img=cv2.imread('/content/1620148797996.png')
h = img.shape[0]
w = img.shape[1]
def xor(img,h,w):
  for iy in range(h):
    for ix in range(w):
      img[iy,ix][2]=int(bin(~img[iy,ix][2]),base=2)
      img[iy,ix][1]=int(bin(~img[iy,ix][1]),base=2)
      img[iy,ix][0]=int(bin(~img[iy,ix][0]),base=2)
  return img
def r_only(img,h,w):
  for iy in range(h):
    for ix in range(w):
      img[iy,ix][1]=0
      img[iy,ix][0]=0
  return img
def g_only(img,h,w):
  for iy in range(h):
    for ix in range(w):
      img[iy,ix][2]=0
      img[iy,ix][0]=0
  return img
def b_only(img,h,w):
  for iy in range(h):
    for ix in range(w):
      img[iy,ix][2]=0
      img[iy,ix][1]=0
  return img
def left_rotate(img,h,w):
  for iy in range(h):
    for ix in range(w):
      img[iy,ix][2]=int(bin(img[iy,ix][2]<<1),base=2)
      img[iy,ix][1]=int(bin(img[iy,ix][1]<<1),base=2)
      img[iy,ix][0]=int(bin(img[iy,ix][0]<<1),base=2)
  return img

cv2_imshow(left_rotate(img,h,w))
```
