```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
img=cv2.imread('/content/1620148797996.png')
h = img.shape[0]
w = img.shape[1]
def xor(img,h,w):
  for iy in range(h):
    for ix in range(w):
      img[iy,ix][2]=int(bin(~img[iy,ix][2]),base=2)
      img[iy,ix][1]=int(bin(~img[iy,ix][1]),base=2)
      img[iy,ix][0]=int(bin(~img[iy,ix][0]),base=2)
  return img
def r_only(img,h,w):
  for iy in range(h):
    for ix in range(w):
      img[iy,ix][1]=0
      img[iy,ix][0]=0
  return img
def g_only(img,h,w):
  for iy in range(h):
    for ix in range(w):
      img[iy,ix][2]=0
      img[iy,ix][0]=0
  return img
def b_only(img,h,w):
  for iy in range(h):
    for ix in range(w):
      img[iy,ix][2]=0
      img[iy,ix][1]=0
  return img
def left_rotate(img,h,w,count):
  for co in range(count):
    for iy in range(h):
      for ix in range(w):
        img[iy,ix][2]=int(bin(img[iy,ix][2]<<1),base=2)+int(bin(img[iy,ix][2])[2],base=2)
        img[iy,ix][1]=int(bin(img[iy,ix][1]<<1),base=2)+int(bin(img[iy,ix][2])[2],base=2)
        img[iy,ix][0]=int(bin(img[iy,ix][0]<<1),base=2)+int(bin(img[iy,ix][2])[2],base=2)
  return img
def r_bit(img,h,w,num):
  img2 = np.zeros( (h, w), np.uint8 )
  for iy in range(h):
    for ix in range(w):
      try:
        if (bin(img[iy,ix][2])[len(bin(img[iy,ix][2]))-num-1])=='1':
          img2[iy,ix]=255
      except IndexError:
        pass
  return img2
def g_bit(img,h,w,num):
  img2 = np.zeros( (h, w), np.uint8 )
  for iy in range(h):
    for ix in range(w):
      try:
        if (bin(img[iy,ix][1])[len(bin(img[iy,ix][1]))-num-1])=='1':
          img2[iy,ix]=255
      except IndexError:
        pass
  return img2
def b_bit(img,h,w,num):
  img2 = np.zeros( (h, w), np.uint8 )
  for iy in range(h):
    for ix in range(w):
      try:
        if (bin(img[iy,ix][0])[len(bin(img[iy,ix][0]))-num-1])=='1':
          img2[iy,ix]=255
      except IndexError:
        pass
  return img2
def half(img,h,w):
  for iy in range(h):
    for ix in range(w):
      img[iy,ix][2]=img[iy,ix][2]*0.5
      img[iy,ix][1]=img[iy,ix][1]*0.5
      img[iy,ix][0]=img[iy,ix][0]*0.5
  return img
def double(img,h,w):
  for iy in range(h):
    for ix in range(w):
      if img[iy,ix][2]>=128:
        img[iy,ix][2]=255
      else:
        img[iy,ix][2]=int(img[iy,ix][2])*2
      if img[iy,ix][1]>=128:
        img[iy,ix][1]=255
      else:
        img[iy,ix][1]=int(img[iy,ix][1])*2
      if img[iy,ix][0]>=128:
        img[iy,ix][0]=255
      else:
        img[iy,ix][0]=int(img[iy,ix][0])*2
  return img
def bit_plane(img,h,w,bit):
  for iy in range(h):
    for ix in range(w):
      try:
        if bin(img[iy,ix][2])[len(bin(img[iy,ix][2]))-bit-1]=='0':
          img[iy,ix][2]=255
        else:
          img[iy,ix][2]=0
      except IndexError:
        pass
      try:
        if bin(img[iy,ix][1])[len(bin(img[iy,ix][1]))-bit-1]=='0':
          img[iy,ix][1]=255
        else:
          img[iy,ix][1]=0
      except IndexError:
        pass
      try:
        if bin(img[iy,ix][0])[len(bin(img[iy,ix][0]))-bit-1]=='0':
          img[iy,ix][0]=255
        else:
          img[iy,ix][0]=0
      except IndexError:
        pass
  return img
cv2_imshow(bit_plane(img,h,w,3))
```
